```python
import numpy as np
import cv2 as cv

%matplotlib inline
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
def fixColor(image):
    return(cv.cvtColor(image, cv.COLOR_BGR2RGB))
```

```python
from IPython.display import Video
Video("Video3.mp4", embed=True)
```

```python
video_stream = cv.VideoCapture('Video3.mp4')

# Randomly select 30 frames
frameIds = video_stream.get(cv.CAP_PROP_FRAME_COUNT) * np.random.uniform(size=30)

# Store selected frames in an array
frames = []
for fid in frameIds:
    video_stream.set(cv.CAP_PROP_POS_FRAMES, fid)
    ret, frame = video_stream.read()
    frames.append(frame)
    
video_stream.release()

```

```python
# Calculate the median along the time axis
medianFrame = np.median(frames, axis=0).astype(dtype=np.uint8)
plt.imshow(fixColor(medianFrame))
```

```python
cv.imwrite('1Claculated_median.png', medianFrame)
```

```python
# Calculate the average along the time axis
avgFrame = np.average(frames, axis=0).astype(dtype=np.uint8)
plt.imshow(fixColor(avgFrame))
```

```python
cv.imwrite('2Claculated_avg.png', avgFrame)
```

```python
writer = cv.VideoWriter("output.mp4", cv.VideoWriter_fourcc(*"MP4V"), 30,(640,480))
```

```python
#Create a new video stream and get total frame count
video_stream = cv.VideoCapture('Video.mp4')
total_frames=video_stream.get(cv.CAP_PROP_FRAME_COUNT)
total_frames
```

```python
frameCnt=0
while(frameCnt < total_frames-1):

    frameCnt+=1
    ret, frame = video_stream.read()

    # Convert current frame to grayscale
    gframe = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    # Calculate absolute difference of current frame and
    # the median frame
    dframe = cv.absdiff(gframe, grayMedianFrame)
    # Gaussian
    blurred = cv.GaussianBlur(dframe, (11, 11), 0)
    #Thresholding to binarise
    ret, tframe= cv.threshold(blurred,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
    #Identifying contours from the threshold
    (cnts, _) = cv.findContours(tframe.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    #For each contour draw the bounding bos
    for cnt in cnts:
        x,y,w,h = cv.boundingRect(cnt)
        if y > 200: # Disregard items in the top of the picture
            cv.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)

    writer.write(cv.resize(frame, (640,480)))
 
#Release video object
video_stream.release()
writer.release()
```
